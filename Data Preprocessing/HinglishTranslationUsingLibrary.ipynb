```python
# Ensure the column exists
if 'crimeaditionalinfo' not in df.columns:
    raise ValueError("The CSV file must contain a 'hinglish_text' column.")
```

```python
# Function to translate text from Hinglish to English
def translate_text(text, row_number):
    global translated_count
    try:
        translated_text = GoogleTranslator(source='auto', target='en').translate(text)
        if translated_text != text:
            print(f"✅ Row {row_number} Translated: {text} -> {translated_text}")
            translated_count += 1
        else:
            print(f"⚠️ Row {row_number} No change: {text}")
        return translated_text
    except Exception as e:
        print(f"❌ Error in Row {row_number} translating text: {text}, Error: {e}")
        return text  # Return original text in case of error
```

```python
# Apply translation and store in a new column
df['translated_text'] = df['crimeaditionalinfo'].apply(lambda x: translate_text(str(x),df.index[df['crimeaditionalinfo'] == x][0]))
```

```python
# Save updated CSV
df.to_csv('/content/output.csv', index=False)
print("CSV file has been updated successfully!")
```

```python
import pandas as pd
from transformers import MarianMTModel, MarianTokenizer
from tqdm import tqdm  # Import tqdm for progress tracking

# Load the MarianMT model for Hindi-English
model_name = "Helsinki-NLP/opus-mt-hi-en"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate_text(text):
    if pd.isna(text) or text.strip() == "":
        return ""  # Handle empty or NaN values
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    translated = model.generate(**inputs)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# Load your CSV file
df = pd.read_csv("/content/train.csv")

# Create a progress bar and translate text in the "crimeaditionalinfo" column
tqdm.pandas(desc="Translating Hinglish to English")
df["English_Translation"] = df["crimeaditionalinfo"].progress_apply(translate_text)

# Save the translated file
output_file = "translated_output.csv"
df.to_csv(output_file, index=False)

print(f"Translation complete! Check '{output_file}'")
```
